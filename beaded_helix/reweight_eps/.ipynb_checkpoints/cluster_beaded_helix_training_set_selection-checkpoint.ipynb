## Choose Training Set Using Poisson Random Number

```python
def poisson_train_predict_sets(total_traj_data, frame_weights, training_set_size, n_training_sets=1):
    # read meta data from input
    n_total_frames = total_traj_data.shape[0]
    n_atoms = total_traj_data.shape[1]
    n_dims = total_traj_data.shape[2]
    # declare arrays
    #training_data = np.empty((n_training_sets,training_set_size,n_atoms,n_dims))
    #cv_data = np.empty((n_training_sets,n_total_frames-training_set_size,n_atoms,n_dims))
    training_data_list = []
    training_frame_weights_list = []
    cv_data_list = []
    cv_frame_weights_list = []
    #
    for set in range(n_training_sets):
        poisson_random_ints = np.random.poisson(lam = training_set_size*frame_weights)
        n_train_frames = np.count_nonzero(poisson_random_ints)
        training_data = np.empty((n_train_frames,n_atoms,n_dims))
        training_frame_weights = np.empty((n_train_frames))
        cv_data = np.empty((n_total_frames-n_train_frames,n_atoms,n_dims))
        cv_frame_weights = np.empty((n_total_frames-n_train_frames,n_atoms,n_dims))
        train_count = 0
        cv_count = 0
        for i in range(n_total_frames):
            if poisson_random_ints[i] > 0:
                training_data[train_count] = total_traj_data[i]
                training_frame_weights[train_count] = frame_weights[i]
                train_count += 1
            else:
                cv_data[cv_count] = total_traj_data[i]
                cv_frame_weights[cv_count] = frame_weights[i]
                cv_count += 1
        training_data_list.append(training_data)
        training_frame_weights_list.append(training_frame_weights)
        cv_data_list.append(cv_data)
        cv_frame_weights_list.append(cv_frame_weights)
    return training_data_list, training_frame_weights_list, cv_data_list, cv_frame_weights_list

def uniform_train_predict_sets(total_traj_data, frame_weights, training_set_size, n_training_sets=1):
    # read meta data from input
    n_total_frames = total_traj_data.shape[0]
    n_atoms = total_traj_data.shape[1]
    n_dims = total_traj_data.shape[2]
    # declare arrays
    #training_data = np.empty((n_training_sets,training_set_size,n_atoms,n_dims))
    #cv_data = np.empty((n_training_sets,n_total_frames-training_set_size,n_atoms,n_dims))
    training_data_list = []
    training_frame_weights_list = []
    cv_data_list = []
    cv_frame_weights_list = []
    #
    index_array = np.arange(n_total_frames).astype(int)
    for set in range(n_training_sets):
        # shuffle trajectory data
        np.random.shuffle(index_array)
        n_train_frames = training_set_size
        # declare arrays
        training_data = np.empty((n_train_frames,n_atoms,n_dims))
        training_frame_weights = np.empty((n_train_frames))
        cv_data = np.empty((n_total_frames-n_train_frames,n_atoms,n_dims))
        cv_frame_weights = np.empty((n_total_frames-n_train_frames,n_atoms,n_dims))
        # populate
        training_data = total_traj_data[index_array[:n_train_frames]]
        training_frame_weights = frame_weights[index_array[:n_train_frames]]
        cv_data = total_traj_data[index_array[n_train_frames:]]
        cv_frame_weights = frame_weights[index_array[n_train_frames:]]
        # add to lists
        training_data_list.append(training_data)
        training_frame_weights_list.append(training_frame_weights)
        cv_data_list.append(cv_data)
        cv_frame_weights_list.append(cv_frame_weights)
    return training_data_list, training_frame_weights_list, cv_data_list, cv_frame_weights_list

training_data_list, training_frame_weights_list, cv_data_list, cv_frame_weights_list = poisson_train_predict_sets(traj_data, frame_weights, 1000)
training_data_list, training_frame_weights_list, cv_data_list, cv_frame_weights_list = uniform_train_predict_sets(traj_data, frame_weights, 1000)
```

```python
def stratify_train_predict_sets(total_traj_data, frame_weights, training_set_size, n_training_sets=1, core_fraction = 0.1):
    # read meta data from input
    n_total_frames = total_traj_data.shape[0]
    n_atoms = total_traj_data.shape[1]
    n_dims = total_traj_data.shape[2]
    # declare arrays
    #training_data = np.empty((n_training_sets,training_set_size,n_atoms,n_dims))
    #cv_data = np.empty((n_training_sets,n_total_frames-training_set_size,n_atoms,n_dims))
    training_data_list = []
    training_frame_weights_list = []
    cv_data_list = []
    cv_frame_weights_list = []
    # sort the weights
    frame_sorted_indeces = np.argsort(frame_weights)[::-1]
    core_size = int(core_fraction * training_set_size)
    remainder = training_set_size - core_size
    # every training set will have half of it populated by the most heavily weighted frames
    fixed_indeces = frame_sorted_indeces[:core_size]
    # the remaining will be uniformly randomly sampled from the remaining frame weights
    shuffle_indeces = frame_sorted_indeces[core_size:]
    # keep an array of all indeces
    index_array = np.arange(n_total_frames).astype(int)
    for set in range(n_training_sets):
        # shuffle trajectory data
        np.random.shuffle(shuffle_indeces)
        train_indeces = np.append(fixed_indeces,shuffle_indeces[:remainder])
        cv_indeces = np.delete(index_array,train_indeces)
        # populate
        training_data = total_traj_data[train_indeces]
        training_frame_weights = frame_weights[train_indeces]
        cv_data = total_traj_data[cv_indeces]
        cv_frame_weights = frame_weights[cv_indeces]
        # add to lists
        training_data_list.append(training_data)
        training_frame_weights_list.append(training_frame_weights)
        cv_data_list.append(cv_data)
        cv_frame_weights_list.append(cv_frame_weights)
    return training_data_list, training_frame_weights_list, cv_data_list, cv_frame_weights_list

def mc_train_predict_sets(total_traj_data, frame_weights, training_set_size, n_training_sets=1, n_mc=10):
    # read meta data from input
    n_total_frames = total_traj_data.shape[0]
    n_atoms = total_traj_data.shape[1]
    n_dims = total_traj_data.shape[2]
    # declare arrays
    #training_data = np.empty((n_training_sets,training_set_size,n_atoms,n_dims))
    #cv_data = np.empty((n_training_sets,n_total_frames-training_set_size,n_atoms,n_dims))
    training_data_list = []
    training_frame_weights_list = []
    cv_data_list = []
    cv_frame_weights_list = []
    #
    index_array = np.arange(n_total_frames).astype(int)
    shuffle_array = np.arange(n_total_frames).astype(int)
    for set in range(n_training_sets):
        # shuffle trajectory data
        np.random.shuffle(shuffle_array)
        train_indeces = shuffle_array[:training_set_size] 
        # populate initial training weights
        training_frame_weights = frame_weights[train_indeces]
        # perform MC
        for mc_step in range(n_mc):
            # shuffle trajectory data
            remaining_indeces = np.delete(index_array,train_indeces)
            np.random.shuffle(remaining_indeces)
            # check to see if we should replace any frames based on Metropolis criteria
            r = np.random.rand(training_set_size)
            for i in range(training_set_size):
                if r[i] <= frame_weights[remaining_indeces[i%remaining_indeces.size]]/frame_weights[train_indeces[i]]:
                    train_indeces[i] = remaining_indeces[i%remaining_indeces.size]
        # populate arrays
        training_data = total_traj_data[train_indeces]
        training_frame_weights = frame_weights[train_indeces]
        cv_indeces = np.delete(index_array,train_indeces)
        cv_data = total_traj_data[cv_indeces]
        cv_frame_weights = frame_weights[cv_indeces]
        # add to lists
        training_data_list.append(training_data)
        training_frame_weights_list.append(training_frame_weights)
        cv_data_list.append(cv_data)
        cv_frame_weights_list.append(cv_frame_weights)
    return training_data_list, training_frame_weights_list, cv_data_list, cv_frame_weights_list
training_data_list, training_frame_weights_list, cv_data_list, cv_frame_weights_list = mc_train_predict_sets(traj_data, frame_weights, 1000)
```

```python
from shapeGMMTorch import scripts
from shapeGMMTorch import similarities
training_set_sizes = np.array([500,1000,2000,3000,4000,5000,6000,7000,8000,10000,20000,50000,75000,100000])
n_training_sets_per_size = 3
n_attempts_per_set = 5
n_samples_per_js = 10000
uniform_size = []
poisson_size = []
uniform_ll = []
poisson_ll = []
uniform_js = []
poisson_js = []
mc_3_size = []
mc_3_ll = []
mc_3_js = []
for i, training_set_size in enumerate(training_set_sizes):
    uniform_training_data, uniform_training_frame_weights, uniform_cv_data, uniform_cv_frame_weights = uniform_train_predict_sets(traj_data, frame_weights, training_set_size,n_training_sets_per_size)
    poisson_training_data, poisson_training_frame_weights, poisson_cv_data, poisson_cv_frame_weights = poisson_train_predict_sets(traj_data, frame_weights, training_set_size,n_training_sets_per_size)
    mc_training_data, mc_training_frame_weights = mc_train_predict_sets(traj_data, frame_weights, training_set_size,n_training_sets_per_size,n_mc=3)[:2]
    for j in range(n_training_sets_per_size):
        uniform_sgmm = scripts.sgmm_fit_with_attempts(uniform_training_data[j], 2, n_attempts_per_set, uniform_training_frame_weights[j],device=torch.device("cpu"))
        uniform_size.append(uniform_training_data[j].shape[0])
        uniform_ll.append(uniform_sgmm.predict(traj_data,frame_weights = frame_weights)[2])
        uniform_js.append(similarities.js_divergence(uniform_sgmm,sgmm_eps8_gt,n_samples_per_js))
        # Poisson
        poisson_sgmm = scripts.sgmm_fit_with_attempts(poisson_training_data[j], 2, n_attempts_per_set, poisson_training_frame_weights[j],device=torch.device("cpu"))
        poisson_size.append(poisson_training_data[j].shape[0])
        poisson_ll.append(poisson_sgmm.predict(traj_data,frame_weights = frame_weights)[2])
        poisson_js.append(similarities.js_divergence(poisson_sgmm,sgmm_eps8_gt,n_samples_per_js))
        #MC
        mc_sgmm = scripts.sgmm_fit_with_attempts(mc_training_data[j], 2, n_attempts_per_set, mc_training_frame_weights[j],device=torch.device("cpu"))
        mc_3_size.append(mc_training_data[j].shape[0])
        mc_3_ll.append(mc_sgmm.predict(traj_data,frame_weights = frame_weights)[2])
        mc_3_js.append(similarities.js_divergence(mc_sgmm,sgmm_eps8_gt,n_samples_per_js))
```

```python
## from scipy import stats
uniform_training_set_sizes = np.unique(uniform_size[:-1])
uniform_ll_array = np.array(uniform_ll[:-1]).reshape((-1,3))
uniform_ll_means = np.mean(uniform_ll_array,axis=1)
uniform_ll_errors = stats.sem(uniform_ll_array,axis=1)
uniform_js_array = np.array(uniform_js[:-1]).reshape((-1,2))
uniform_js_array = uniform_js_array[:,0].reshape(-1,3)
uniform_js_means = np.mean(uniform_js_array,axis=1)
uniform_js_errors = stats.sem(uniform_js_array,axis=1)
mc_training_set_sizes = np.unique(mc_3_size)
mc_3_ll_array = np.array(mc_3_ll).reshape((-1,3))
mc_ll_means = np.mean(mc_3_ll_array,axis=1)
mc_ll_errors = stats.sem(mc_3_ll_array,axis=1)
mc_js_array = np.array(mc_3_js).reshape((-1,2))
mc_js_array = mc_js_array[:,0].reshape(-1,3)
mc_js_means = np.mean(mc_js_array,axis=1)
mc_js_errors = stats.sem(mc_js_array,axis=1)
poisson_training_set_sizes = np.mean(np.array(poisson_size[:-1]).reshape(-1,3),axis=1)
poisson_x_error = stats.sem(np.array(poisson_size[:-1]).reshape(-1,3),axis=1)
poisson_ll_array = np.array(poisson_ll[:-1]).reshape((-1,3))
poisson_ll_means = np.mean(poisson_ll_array,axis=1)
poisson_ll_errors = stats.sem(poisson_ll_array,axis=1)
poisson_js_array = np.array(poisson_js[:-1]).reshape((-1,2))
poisson_js_array = poisson_js_array[:,0].reshape(-1,3)
poisson_js_means = np.mean(poisson_js_array,axis=1)
poisson_js_errors = stats.sem(poisson_js_array,axis=1)
```
